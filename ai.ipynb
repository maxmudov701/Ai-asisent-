```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_uzbekistan():
    fig, ax = plt.subplots(figsize=(9, 4.5))

    # Ranglar
    ax.add_patch(patches.Rectangle((0, 0.66), 1, 0.33, color='#0099B5')) # Moviy
    ax.add_patch(patches.Rectangle((0, 0.33), 1, 0.33, color='white'))   # Oq
    ax.add_patch(patches.Rectangle((0, 0), 1, 0.33, color='#1EB53A'))    # Yashil

    # Qizil hoshiyalar
    ax.add_patch(patches.Rectangle((0, 0.65), 1, 0.02, color='#CE1126'))
    ax.add_patch(patches.Rectangle((0, 0.32), 1, 0.02, color='#CE1126'))

    # Yarim oy
    moon = plt.Circle((0.15, 0.82), 0.1, color='white')
    moon_cut = plt.Circle((0.18, 0.82), 0.09, color='#0099B5')
    ax.add_artist(moon)
    ax.add_artist(moon_cut)

    # Yulduzlar (soddalashtirilgan nuqtalar)
    for i in range(3): # Birinchi qator
        ax.plot(0.28 + i*0.04, 0.88, 'w*', markersize=5)
    for i in range(4): # Ikkinchi qator
        ax.plot(0.28 + i*0.04, 0.82, 'w*', markersize=5)
    for i in range(5): # Uchinchi qator
        ax.plot(0.28 + i*0.04, 0.76, 'w*', markersize=5)

    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')
    plt.title("O'zbekiston bayrog'i")
    plt.show()

draw_uzbekistan()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_france():
    fig, ax = plt.subplots(figsize=(9, 6))

    # Ranglar: Ko'k, Oq, Qizil
    ax.add_patch(patches.Rectangle((0, 0), 1/3, 1, color='#002395'))
    ax.add_patch(patches.Rectangle((1/3, 0), 1/3, 1, color='white'))
    ax.add_patch(patches.Rectangle((2/3, 0), 1/3, 1, color='#ED2939'))

    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')
    plt.title("Fransiya bayrog'i")
    plt.show()

draw_france()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_thailand():
    fig, ax = plt.subplots(figsize=(9, 6))

    # Ranglar va balandliklar
    # Qizil (yuqori)
    ax.add_patch(patches.Rectangle((0, 5/6), 1, 1/6, color='#ED1C24'))
    # Oq
    ax.add_patch(patches.Rectangle((0, 4/6), 1, 1/6, color='white'))
    # To'q ko'k (o'rtadagi qalin chiziq)
    ax.add_patch(patches.Rectangle((0, 2/6), 1, 2/6, color='#241D4F'))
    # Oq
    ax.add_patch(patches.Rectangle((0, 1/6), 1, 1/6, color='white'))
    # Qizil (pastki)
    ax.add_patch(patches.Rectangle((0, 0), 1, 1/6, color='#ED1C24'))

    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')
    plt.title("Tailand bayrog'i")
    plt.show()

draw_thailand()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_japan():
    # Bayroq o'lchami 2:3 nisbatda
    fig, ax = plt.subplots(figsize=(9, 6))

    # Oq fon (asosiy mato)
    ax.add_patch(patches.Rectangle((0, 0), 1, 1, color='white', ec='black', lw=1))

    # Qizil doira (Quyosh) markazda
    # Markaz (0.5, 0.5), radiusi 0.3
    sun = plt.Circle((0.5, 0.5), 0.3, color='#BC002D')
    ax.add_artist(sun)

    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal') # Doira aylana shaklida qolishi uchun
    ax.axis('off')
    plt.title("Yaponiya bayrog'i")
    plt.show()

draw_japan()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_indonesia():
    # Bayroq o'lchami 2:3 nisbatda
    fig, ax = plt.subplots(figsize=(9, 6))

    # Yuqori qismi: Qizil
    ax.add_patch(patches.Rectangle((0, 0.5), 1, 0.5, color='#FF0000'))

    # Pastki qismi: Oq
    ax.add_patch(patches.Rectangle((0, 0), 1, 0.5, color='white', ec='lightgray', lw=0.5))

    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')
    # Ramka qo'shish (oq qismi ko'rinishi uchun)
    for spine in ax.spines.values():
        spine.set_visible(True)

    plt.title("Indoneziya bayrog'i")
    plt.show()

draw_indonesia()
```
